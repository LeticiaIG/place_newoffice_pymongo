```python
def RemoveWhiteSpaces(row):
    row = str(row)
    row = row.replace(" ","")
    return row
df['total_money_raised'] = df['total_money_raised'].apply(RemoveWhiteSpaces)
```

```python
df['currency'] = df['total_money_raised'].str[0]
df['total_money_raised'] = df['total_money_raised'].str[1:]
```

```python
def ChangeNonNumeric(row):
    row = str(row)
    try:
        if "B" in row:
            row = row.replace("B", "")
            row = float(row)*1000000000
        elif "M" in row:
            row = row.replace("M", "")
            row = float(row)*1000000
            
            
            
        elif "k" in row:
            row = row.replace("k", "")
            row = float(row)*1000
    except: row=None
    return row
```

```python
df['total_money_raised'] = df['total_money_raised'].apply(ChangeNonNumeric)
df
```

```python
def change_string(row):
    row = str(row)
    return row
def merge_currency_columns(df):
    df['currency'] = df['currency'].apply(change_string)
    df['total_money_raised'] = df['total_money_raised'].apply(change_string)
    df['currency_n'] = df['total_money_raised'] + df['currency']
    return df
df=merge_currency_columns(df)
```

```python
url = 'https://api.exchangerate-api.com/v4/latest/USD'
res = requests.get(url)
data = res.json()
    # EXCHANGE RATE API
def exchangeRate (row):
    row = str(row)
    try:
        if "$" in row:
            row = row.replace("$","")
            row = float(row)*data['rates']['USD']
        if "€" in row:
            row = row.replace("€","")
            row = float(row)*data['rates']['EUR']
        if "£" in row:
            row = row.replace("£","")
            row = float(row)*data['rates']['GBP']
        else:
            row = row.replace("$","")
            row = float(row)*data['rates']['USD']     
    except:
        row = row
    return row
def Apply_exchangeRate(df):
    df['currency_n'] = df['currency_n'].apply(exchangeRate)
    return df
df1_d = Apply_exchangeRate(df)
print(df1_d)
```

```python
import googlemaps
import pprint
import time
import requests
import os
from dotenv import load_dotenv
load_dotenv()
key = os.getenv("G_CLOUD_API_KEY")

# GOOGLEPLACES --- 
def get_lat_long_places(name):
    try:
        query = "{} headquarter".format(name)
        print(query)
        url = "https://maps.googleapis.com/maps/api/place/textsearch/json?query={}&key={}".format(query,key)
        response = requests.get(url)
        data = response.json()
        location = data.get('results')[0].get('geometry').get('location')
        return{
            "latitud" : location.get('lat'),
            "longitud" : location.get('lng')
        }
    except: 
        return {
            "latitud" : None,
            "longitud" : None
        }    
#data['results'][0]
```
